### Проверка эффективности обогащения

Формула для расчёта эффективности обогащения: `Recovery = (C * (F - T)) / (F * (C - T))`  
Для расчёта из обучающей выборки понадобятся столбцы:  
C - `rougher.output.concentrate_au`,  

F - `rougher.input.feed_au`,  

T - `rougher.output.tail_au`.

```python
C = gr_train['rougher.output.concentrate_au']
F = gr_train['rougher.input.feed_au']
T = gr_train['rougher.output.tail_au']
```

```python
rougher_output_recovery_recalculated = (C * (F - T)) / (F * (C - T)) * 100
```

```python
mean_absolute_error(gr_train[['rougher.output.recovery']], rougher_output_recovery_recalculated)
```

Значение средней абсолютной ошибки очень низкое, что говорит о корректном расчёте эффективности обогащения в столбце `rougher.output.recovery`.

```python
def drop_anomalies(data):
    anomalous_columns = [col for col in data.columns if (col.find('output.') != -1 and
                                                         col.find('output.tail') == -1 or
                                                         col.find('input.feed') != -1)]
    data[anomalous_columns] = data[anomalous_columns][data[anomalous_columns] >= 1]
    return(data)
```

## Модель

### Функция для вычисления итоговой sMAPE

sMAPE рассчитывается как сумма значений модуля разности таргета и предикта, делённого на сумму модулей таргета и предикта, делённую пополам, делённая на количество значений. Выражается в процентном значении.

```python
def smape(target, pred):
    smape = abs(target - pred) / ((abs(target) + abs(pred))/2) * 100
    smape = smape.mean()
    return(smape)
```

```python
def smape_final(smape_r, smape_f):
    smape_final = .25 * smape_r + .75 * smape_f
    return(smape_final)
```

### Обучение моделей и выбор лучшей

Перед обучением проведём масштабирование признаков.

```python
def scaler(features_train, features_test):
    scaler = StandardScaler()
    scaler.fit(features_train)
    columns = features_test.columns
    features_train[columns] = scaler.transform(features_train[columns])
    features_test[columns] = scaler.transform(features_test[columns])
```

```python
scaler(features_train_r, features_test_r)
scaler(features_train_f, features_test_f)
```

```python
def dt_model(features_train, target_train, \
             depth_range, samples_leaf_range):
    best_smape = 100
    for depth in depth_range:
        for samples_leaf in samples_leaf_range:
            model = DecisionTreeRegressor(max_depth=depth, min_samples_leaf=samples_leaf, random_state=RANDOM_STATE)
            model.fit(features_train, target_train)
            smapes = cross_val_score(model, features_train, target_train, cv=5, scoring=make_scorer(smape, greater_is_better=False))
            final_smape = sum(-smapes) / len(smapes)
            if final_smape < best_smape:
                best_smape = final_smape
                best_depth = depth
                best_samples_leaf = samples_leaf
                best_dt_model = model
    print(f'Самый низкий показатель sMAPE = {best_smape} при гиперпараметрах: ' + \
          f'max_depth = {best_depth}, ' + \
          f'min_samples_leaf = {best_samples_leaf}')
    return(best_dt_model)
```

```python
def rf_model(features_train, target_train, \
             depth_range, n_est_range):
    best_smape = 100
    for depth in depth_range:
        for n_est in n_est_range:
            model = RandomForestRegressor(max_depth=depth, n_estimators=n_est, random_state=RANDOM_STATE)
            model.fit(features_train, target_train)
            smapes = cross_val_score(model, features_train, target_train, cv=5, scoring=make_scorer(smape, greater_is_better=False))
            final_smape = sum(-smapes) / len(smapes)
            if final_smape < best_smape:
                best_smape = final_smape
                best_depth = depth
                best_n_est = n_est
                best_rf_model = model
    print(f'Самый низкий показатель sMAPE = {best_smape} при гиперпараметрах: ' + \
          f'max_depth = {best_depth}, ' + \
          f'n_estimators = {best_n_est}')
    return(best_rf_model)
```

Обучим разные модели и найдём гиперпараметры, при которых метрика smape будет ближе к нулю. Определять качество модели будем посредством кросс валидации.  
Начнём с модели регрессии деревом решений.

```python
%%time
dt_model_r = dt_model(features_train_r, target_train_r, \
                      range(1, 15, 2), range(1, 502, 50))
```

```python
%%time
dt_model_f = dt_model(features_train_f, target_train_f, \
                      range(1, 15, 2), range(1, 502, 50))
```

```python
%%time
rf_model_r = rf_model(features_train_r, target_train_r, \
                      range(15, 26, 5), range(60, 71, 10))
```

```python
%%time
rf_model_f = rf_model(features_train_f, target_train_f, \
                      range(10, 21, 5), range(90, 91, 10))
```

```python
lr_model_r = LinearRegression()
lr_model_r.fit(features_train_r, target_train_r)
scores = cross_val_score(lr_model_r, features_train_r, target_train_r, cv=5, scoring=make_scorer(smape, greater_is_better=False))
final_score = sum(-scores) / len(scores)
print('показатель sMAPE =', final_score)
```

```python
lr_model_f = LinearRegression()
lr_model_f.fit(features_train_f, target_train_f)
scores = cross_val_score(lr_model_f, features_train_f, target_train_f, cv=5, scoring=make_scorer(smape, greater_is_better=False))
final_score = sum(-scores) / len(scores)
print('показатель sMAPE =', final_score)
```

Наименьшие показатели sMAPE при кросс-валидации показывают регрессионные модели деревом решений.

Проверим адекватность модели. Сравним оценку sMAPE с оценкой модели, которая даёт все предикты медианным значением.

```python
def dummy(features_train, target_train):
    dummy_model = DummyRegressor(strategy='median')
    dummy_model.fit(features_train, target_train)
    dummy_predict = dummy_model.predict(features_train)
    return(smape(target_train, dummy_predict))
```

```python
dummy(features_train_r, target_train_r)
```

```python
dummy(features_train_f, target_train_f)
```

sMAPE дамми модели для таргета эффективности обогащения чернового концентрата выше, чем у всех обученных моделей.  
sMAPE дамми модели для таргета эффективности обогащения финального концентрата выше, чем у моделей деревом решений и случайный лес, но ниже, чем у модели линейной регрессии.

Финальное тестирование проведём с моделями деревом решений.

```python
def test_model(model, features_test, target_test):
    predictions = model.predict(features_test)
    smape_ = smape(target_test, predictions)
    print('Качество модели:', smape_)
    return(smape_)
```

```python
dt_smape_r = test_model(dt_model_r, features_test_r, target_test_r)
dt_smape_f = test_model(dt_model_f, features_test_f, target_test_f)
```

```python
final_smape = smape_final(dt_smape_r, dt_smape_f)
print('Итоговое значение sMAPE при работе регрессионных моделей деревом решений на тестовой выборке =', final_smape)
```
