**1. Общая площадь и количество комнат**.

В столбцах `total_area` и `rooms` нет пропусков, но максимальные значения слишком высокие. Проверим столбцы на артефакты, постараемся понять их природу и исправить, в противном случае избавимся от строк.  

Рассмотрим, как распределяются квартиры по количеству комнат.

```python
data.groupby('rooms')['rooms'].count()
```

В столбце `rooms` есть значение 0, то есть количество комнат в квартире равно нулю. Если значение столбца `studio` True, тогда это студия, если же значение столбца `open_plan` True, тогда это квартира со свободной планировкой. В остальных случаях нужно удалить эту строку.  
Максимальное количество комнат в квартире возьмём 5, максимальная площадь 150 квадратных метров.

Большая часть данных включает диапазон комнат от 0 до 5, а следовательно, все строки, где количество комнат более 5, можно удалить.

```python
data = data.query('((1 <= rooms <= 5 and studio == False and open_plan == False) or\
                  (rooms == 0 and (studio == True or open_plan == True))) and\
                  total_area <= 150')
```

**2. Высота потолков.**

Рассмотрим столбец `ceiling_height`. Устраивают данные только по тем квартирам, высота потолков в которых находится в диапазоне от 2 до 6 метров, квартиры могут быть низкими или же двухэтажными. Пропуски заполним медианным значением, но перед этим разберёмся с артефактами. Если в таблице есть значения от 20 до 60, значит их могли неправильно записать, возможно, истинное значения должно было быть с десятичной дробью, например не 35 метров, а 3.5 метров.

```python
# Цикл с условием обрабатывают строки, в которых значение высоты потолка от 20 до 60 метров, то есть аномальное,
# разделяя его на 10.
for i in data.index:
    if 20 <= data.loc[i, 'ceiling_height'] <= 60:
        data.loc[i, 'ceiling_height'] = data.loc[i, 'ceiling_height'] / 10

# Методом .query() избавляемся от строк, у которых значение высоты потолка менее 2 и более 6.
data = data.query('2 <= ceiling_height <= 6 or ceiling_height.isna() == True')

# Пустые значения заполняем медианным по столбцу 'ceiling_height'
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

**3. Этажность дома.**

Пустые значения в столбце `floors_total` попытаемся заполнить медианным значением, где это возможно. Если же этаж квартиры больше медианного значения столбца этажности дома, тогда присвоим значению этажности дома этаж квартиры.

```python
# Пустой список, который далее заполняется индексами строк, где в столбце floors_total пустое значение.
floors_total_nan_id = []
floors_total_nan_id.extend(data.query('floors_total.isna()').index)

#медианное значение столбца floors_total
floors_total_median = data['floors_total'].median()

# Цикл, заполняющий пустые значения медианным, если в этой строке значение floor менее значения медианы floors_total_median,
# иначе присваивает этажности дома значение этажа, на котором расположена квартира.
for i in floors_total_nan_id:
    if (data.loc[i, 'floor'] < floors_total_median):
        data.loc[i, 'floors_total'] = floors_total_median
    else:
        data.loc[i, 'floors_total'] = data.loc[i, 'floor']
```

**4. Общая площадь с площадью кухни и жилой площадью.**

Максимальное значение жилой площади `living_area` в датасете 128 кв.м., что допустимо, так как мы рассматриваем квартиры до 150 кв.м. по общей площади. Минимальное значение 2 кв.м., что недопустимо, возьмём минимальное значение жилой площади 10 кв.м.  

Площадь кухни `kitchen_area` возьмём от 0 до 25 квадратных метров. Максимальное значение площади кухни по всему датасету 93 кв.м. Возможно, информацию внесли некорректно, поэтому все значения от 50 до 93 кв.м. поделим на 10.  

Пропуски заполним медианным значением по количеству комнат.  

Общая площадь может быть больше суммы жилой площади и площади кухни, так как ещё может быть площадь подсобных помещений, ванных комнат, но не может быть меньше. Проверим, чтобы сумма жилой площади и площади кухни была не больше общей площади, иначе присвоим общей площади эту сумму.

```python
# Фильтруем датасет по жилой площади, минимальная площадь от 10 кв.м.
data = data.query('living_area >= 10 or living_area.isna() == True')

# Функция проверяет в строке значение площади кухни, если оно в диапазоне от 50 до 93 кв.м., тогда вернуть
# значение, делённое на 10.
def kitchen_area_anomaly(row):
    if 50 <= row['kitchen_area'] <= 93:
        row['kitchen_area'] = row['kitchen_area'] / 10
    return(row['kitchen_area'])

# Методом .apply() применяем функию к каждой строке датафрейма.
data['kitchen_area'] = data.apply(kitchen_area_anomaly, axis=1)

# Фильтруем датафрейм по значению площади кухни от 0 до 25 кв.м.
data = data.query('0 <= kitchen_area <= 25 or kitchen_area.isna()')

# Цикл заполняет пустые значения столбцов 'living_area' и 'kitchen_area' медианными значениями по количеству комнат.
for r in data['rooms'].unique():
    data.loc[(data['rooms'] == r) & (data['living_area'].isna()), 'living_area'] = \
    data.loc[(data['rooms'] == r), 'living_area'].median()
    data.loc[(data['rooms'] == r) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[(data['rooms'] == r), 'kitchen_area'].median()

# Функция, в которой проверяется условие, если общая площадь квартиры менее суммы площади кухни и жилой площади, тогда
# присвоить значению общей площади эту сумму, иначе всё останется без изменений.
def area_sum(row):
    if row['kitchen_area'] + row['living_area'] > row['total_area']:
        row['total_area'] = row['kitchen_area'] + row['living_area']
    return(row['total_area'])

# Применяем функцию к каждой строке датафрейма методом .apply().
data['total_area'] = data.apply(area_sum, axis=1)
```

Проверим, все ли пропуски заполнены в столбцах `living_area` и `kitchen_area`.

```python
data[['living_area', 'kitchen_area']].info()
print(data.groupby('rooms')['kitchen_area'].count())
```

Все пропуски жилой площади заполнены, но остались пропуски по площади кухни. Заполним оставшиеся пропуски нулём, так как в этих квартирах кухни может вообще не быть.

Отфильтруем датасет следующим образом, максимальное количество балконов в студии 1, в однокомнатной квартире 2, в двухкомнатной 3 и так далее. Максимальное количество балконов в квартире по датасету 5, следовательно кваритры с 4 и 5 комнатами могут включать максимум 5 балконов.  

Пропущенные значения заполним нулём, так как владелец квартиры мог не указать информацию о количестве балконов, потому что в квартире вовсе нет ни одного балкона.

```python
data = data.query('rooms == 0 and balcony <= 1 or \
                   rooms == 1 and balcony <= 2 or \
                   rooms == 2 and balcony <= 3 or \
                   rooms == 3 and balcony <= 4 or \
                   rooms in [4, 5] and balcony <= 5 or \
                   balcony.isna()')

data['balcony'] = data['balcony'].fillna(0)
```

**6. Картографические данные и дубликаты.**

Уберём из датафрейма строки, в которых населённый пункт неизвестный. Затем обработаем неявные дубликаты в столбце `locality_name`. Для этого сделаем нижний регистр значений столбца, отсортируем уникальные значения по алфавиту и методом `.str.replace()` приведём строки к общему виду, например, изменим все буквы 'ё' на 'е' и т.д.

```python
data = data.dropna(subset='locality_name')
```

```python
data['locality_name'] = (
    data['locality_name'].str
    .replace('ё', 'е').str
    .replace('поселок городского типа ', '').str
    .replace('деревня ', '').str
    .replace('садовое товарищество ', '').str
    .replace('село ', '').str
    .replace('городской поселок ', '').str
    .replace('поселок станции ', '').str
    .replace('поселок при железнодорожной станции ', '').str
    .replace('коттеджный поселок ', '').str
    .replace('поселок ', '').str
    .replace('красносельское', 'красное село').str
    .replace('саперный', 'саперное')
)
```

После работы с дубликатами можно переходить к заполнению пропусков. Заполнять будем пропуски в столбцах `nearest_airport`, `nearest_city_center`, `parks_around_3_km`, `nearest_park`, `ponds_around_3_km`, `nearest_pond` медианными значениями по населённым пунктам.

```python
# Цикл заполняет пустые значения медианными по населённым пунктам.
for n in data['locality_name'].unique():
    data.loc[(data['locality_name'] == n) & (data['nearest_airport'].isna()), 'nearest_airport'] = \
    data.loc[(data['locality_name'] == n), 'nearest_airport'].median()
    
    data.loc[(data['locality_name'] == n) & (data['nearest_city_center'].isna()), 'nearest_city_center'] = \
    data.loc[(data['locality_name'] == n), 'nearest_city_center'].median()
    
    data.loc[(data['locality_name'] == n) & (data['parks_around_3_km'].isna()), 'parks_around_3_km'] = \
    data.loc[(data['locality_name'] == n), 'parks_around_3_km'].median()
    
    data.loc[(data['locality_name'] == n) & (data['nearest_park'].isna()), 'nearest_park'] = \
    data.loc[(data['locality_name'] == n), 'nearest_park'].median()
    
    data.loc[(data['locality_name'] == n) & (data['ponds_around_3_km'].isna()), 'ponds_around_3_km'] = \
    data.loc[(data['locality_name'] == n), 'ponds_around_3_km'].median()
    
    data.loc[(data['locality_name'] == n) & (data['nearest_pond'].isna()), 'nearest_pond'] = \
    data.loc[(data['locality_name'] == n), 'nearest_pond'].median()
```

Заполнились не все пропуски. Это связано с тем, что во всех строках определённых населённых пунктов отсутствует информация о ближайшем аэропорте, ближайшем парке, ближайшем пруде и т.д. Из-за отсутствия данных не получается рассчитать медиану. Оставшиеся пропуски придётся удалить, так как заполнять их медианным значением по всему столбцу некорректно.

## Расчёт и добавление новых столбцов в таблицу

Необходимо создать 6 новых столбцов.

```python
# Стоимость 1 кв.м.
data['unit_area_price'] = data['last_price'] / data['total_area']
```

```python
# День недели, в который опубликовали объявление. 0 - понедельник, 1 - вторник, 2 - среда и т.д.
data['publication_weekday'] = data['first_day_exposition'].dt.weekday

# Функция, заменяющая численное на строковое название дня недели.
def weekday_func(row):
    if row['publication_weekday'] == 0:
        return('понедельник')
    elif row['publication_weekday'] == 1:
        return('вторник')
    elif row['publication_weekday'] == 2:
        return('среда')
    elif row['publication_weekday'] == 3:
        return('четверг')
    elif row['publication_weekday'] == 4:
        return('пятница')
    elif row['publication_weekday'] == 5:
        return('суббота')
    else:
        return('воскресенье')

# Применяем функцию ко всем строкам.
data['publication_weekday_name'] = data.apply(weekday_func, axis=1)
```

```python
# Месяц, в который опубликовали объявление
data['publication_month'] = data['first_day_exposition'].dt.month

# Функция, заменяющая численное на строковое название месяца.
def month_func(row):
    if row['publication_month'] == 1:
        return('январь')
    elif row['publication_month'] == 2:
        return('февраль')
    elif row['publication_month'] == 3:
        return('март')
    elif row['publication_month'] == 4:
        return('апрель')
    elif row['publication_month'] == 5:
        return('май')
    elif row['publication_month'] == 6:
        return('июнь')
    elif row['publication_month'] == 7:
        return('июль')
    elif row['publication_month'] == 8:
        return('август')
    elif row['publication_month'] == 9:
        return('сентябрь')
    elif row['publication_month'] == 10:
        return('октябрь')
    elif row['publication_month'] == 11:
        return('ноябрь')
    else:
        return('декабрь')

# Применяем функцию ко всем строкам.
data['publication_month_name'] = data.apply(month_func, axis=1)
```

```python
# Год, в который опубликовали объявление
data['publication_year'] = pd.DatetimeIndex(data['first_day_exposition']).year
```

```python
# Функция определяет, какой тип этажа у квартиры.
def floor_type_func(row):
    if row['floor'] == 1:
        return('первый')
    elif row['floor'] == row['floors_total']:
        return('последний')
    return('другой')

# Применяем функцию к каждой строке датафрейма и создаём необходимый столбец.
data['floor_type'] = data.apply(floor_type_func, axis=1)
```

```python
# Создаём столбец, в котором хранится растояние до центра города в км.
data['city_center_distance_km'] = data['nearest_city_center'] / 1000

# Округляем полученное значение и меняем тип данных с float на integer.
data['city_center_distance_km'] = round(data['city_center_distance_km']).astype('int64')
```

Также рассмотрим, как менялась цена квартир с разным количеством комнат по годам.

```python
# Сгруппируем таблицу по году публикации и количеству комнат в квартире.
years_rooms_group = data.groupby(['publication_year', 'rooms'])['last_price'].agg(['median', 'mean'])
display(years_rooms_group)
```

```python
# Функция заполняет словарь с годами в ключах значениями по указанному
# столбцу (медианная или средняя стоимость) по количеству комнат в квартире
# и строит график.
def graphic(column, title):
    plt.figure(figsize=(10, 10))
    x = range(2014, 2020)
    f_rooms = 0
    values = {2014: None, 2015: None, 2016: None, 2017: None, 2018: None, 2019: None}
    for year in x:
        for i in range(len(years_rooms_group.index)):
            if years_rooms_group.index[i][1] == f_rooms:
                values.update({years_rooms_group.index[i][0]: years_rooms_group[column].values[i]})
        plt.plot(x, values.values(), '-o')
        values = {2014: None, 2015: None, 2016: None, 2017: None, 2018: None, 2019: None}
        f_rooms += 1
    plt.legend(range(0,6))
    plt.grid()
    plt.xlabel('Год')
    plt.ylabel('Стоимость')
    plt.title('График ' + title + ' стоимости квартир с разным количеством комнат по годам')
```

```python
graphic('median', 'медианной')
graphic('mean', 'средней')
```

Обработаны все столбцы, содержащие пропущенные значения. Пропуски заполнялись медианными значениями, где это было возможно. Если пропуски не удавалось заполнить, эти строки удалялись.  
3. Изменение типов данных.  
В большинстве столбцов тип данных меняется на целочисленный integer, так как изначально значения записаны целым числом, но их тип float. Например, тип столбца `balcony` или `floors_total` не может быть числом с десятичной дробью, это только целое число. 
Тип столбца `first_day_exposition` переведён из строкового object в datetime, так как это формат даты и времени.  
Тип столбца `is_apartment` переведён в boolean, так как значения в столбце только True или False.
4. Дубликаты.  
По столбцу `locality_name` значения были приведены в нижний регистр, значения скорректированы так, чтобы не было неявных дубликатов, например 'посёлок' и 'поселок'. Явных дубликатов в датафрейме не выявлено.
